```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
%matplotlib inline
```

```python
iris = load_iris()
ir = pd.DataFrame(iris.data, columns=iris.feature_names)
ir.head()
```

```python
print(iris.DESCR)
```

```python
ir[ir.isnull().any(axis=1)]
```

```python
ir.describe()
```

```python
sns.heatmap(ir.corr(), annot=True, fmt='.3f')
sns.heatmap(ir.corr())
```

```python
fig, ax = plt.subplots(figsize=(10,10)) 
sns.distplot(ir['sepal length (cm)'])
```

```python
sns.jointplot(x='sepal length (cm)', y='sepal width (cm)', data=ir, kind="hex")
```
